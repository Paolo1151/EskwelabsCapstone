```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup

import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np

import os
import re
import time
import random

#######################
# Config

HOME_URL = 'https://www.google.com/'

WEBDRIVER = os.getcwd() + '\\chromedriver\\chromedriver.exe'

CLEANED_DATA_PATH = os.getcwd() + '\\..\\..\\data\\cleaned\\PlanetSystemClean.csv'

#######################


def main():
    driver = webdriver.Chrome(WEBDRIVER)
    driver.get(HOME_URL)

    df = pd.read_csv(CLEANED_DATA_PATH)

    planets = list(df['PlanetName'].unique())

    planet_types = []
    for i, planet in enumerate(planets):
        searchbox = driver.find_element_by_name('q')

        searchbox.clear()
        searchbox.send_keys(f'{planet} planet type')
        searchbox.send_keys(Keys.RETURN)

        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        val = soup.find_all('b')

        if val:
            val = val[0].get_text()
            if re.match('.*exoplanet.*', val):
                planet_types.append(val)
            else:
                val = 'NEED MANUAL'
                planet_types.append(val)                  
        else:
            val = 'NEED MANUAL'
            planet_types.append(val)

        print(f"{planet} is a: {val}")
        
        time.sleep(random.randint(1,4))
        
    df['PlanetType'] = planet_types

    df.to_csv(os.getcwd() + '\\..\\..\\data\\cleaned\\PlanetSystemLabelled.csv')

    display(df)
    

if __name__ == '__main__':
    main()
```

# NASA SCRAPER

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

import warnings
warnings.filterwarnings('ignore')

import os
import time
import random
import re

###################
# Config

HOME_URL = 'https://exoplanets.nasa.gov/discovery/exoplanet-catalog/'

WEBDRIVER = os.getcwd() + '\\chromedriver\\chromedriver.exe'

CLEANED_DATA_PATH = os.getcwd() + '\\..\\..\\data\\final\\UnfinishedLabelFinal.csv'

LABEL_TBD = 'TBD'

OUTPUT_FILE_NAME = 'UnverifiedPlanetsFinal.csv'

###################

def main():
    driver = webdriver.Chrome(WEBDRIVER)
    driver.get(HOME_URL)
    
    df = pd.read_csv(CLEANED_DATA_PATH)

    oois = list(df['ObjectOfInterest'])

    planet_types = []
    for ooi in oois:
        plt_typ = df[df['ObjectOfInterest']==ooi]['PlanetType'].iloc[0]
        planet = df[df['ObjectOfInterest']==ooi]['PlanetName'].iloc[0]
        if (plt_typ != LABEL_TBD):
            print(f'{planet} is a: {plt_typ}')
            planet_types.append(plt_typ)
            continue
        elif (plt_typ == 'N/A'):
            planet_types.append(plt_typ)
            continue
        elif (planet == 'N/A'):
            planet_types.append(plt_typ)
            continue
        
        searchbox = driver.find_element_by_id('desktop_search_field')
        searchbox.clear()
        searchbox.send_keys(planet)
        searchbox.send_keys(Keys.RETURN)
        
        time.sleep(random.randint(10, 15))
        
        try:
            link = driver.find_element_by_xpath(f'//ul[@class=\'exoplanet\']//a[contains(text(), \'{planet}\')]').get_attribute('href')
            driver.get(link)
            val = driver.find_element_by_xpath('//td[@class=\'first planet_fact\']//div[@class=\'value\']').text
            driver.back()
        except:
            val = 'UNKNOWN'

        planet_types.append(val)

        print(f'{planet} is a: {val}')

        time.sleep(random.randint(1,3))

    df['PlanetType'] = planet_types

    df.to_csv(os.getcwd() + f'\\..\\..\\data\\processed\\{OUTPUT_FILE_NAME}')
        
    display(df)

if __name__ == '__main__':
    main()
```
