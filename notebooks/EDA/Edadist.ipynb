```python
import pandas as pd
import numpy as np

import os
```

```python
data = pd.read_csv(os.getcwd()+'\\..\\..\\data\\final\\UnverifiedLabelFinal.csv')

# Remove Useless Features
data = data.drop(columns=['PlanetInsolationValue'])

# Remove Outliers
data = data[data['StellarSunRadius']<2.5]
data = data[data['PlanetEarthRadius']<20]
data = data[data['StellarEffectiveTemperature']<7000]

data.to_csv('OutlierFree.csv')

data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

from scipy.stats import kstest

numerics = [col for col in data.columns if data[col].dtype!=object]

for num in numerics:
    temp = data[data['Disposition']=='CFRM'][num]

    fig = go.Figure()
    
    fig.add_trace(
        go.Histogram(x=temp)
    )

    fig.update_layout(
        title=f"Distribution of {num} Column", 
        font=dict(
            family='sans serif',
            size=18
        ),
        paper_bgcolor='rgba(4,17,65,0)',
        polar=dict(
            radialaxis=dict(
                visible=False
            )
        )
    )

    fig.show()
    
```
